```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris=load_iris()

X=iris.data
y=iris.target
```

```python
sc=StandardScaler()

X_scaled=sc.fit_transform(X)
```

```python
from sklearn.cluster import DBSCAN

dbscan=DBSCAN(
    eps=0.8,
    min_samples=5
)
```

```python
label=dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,2],c=label)
```

## Non linear data

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(
    n_samples=100,
    noise=0.05,
    random_state=42
)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
from sklearn.cluster import KMeans
k=2
km=KMeans(n_clusters=k)

label=km.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=label)
```

```python
db=DBSCAN(
    eps=0.2,
    min_samples=5
)

label=db.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=label)
```
